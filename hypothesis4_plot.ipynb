```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'model_name': ['ALBERT'] * 6 + ['MobileBERT'] * 6 + ['REMBERT'] * 6,
    'embedding_config': (['Default'] * 3 + ['ConceptNet'] * 3) * 3,
    'gender': ['Neutral', 'Male', 'Female'] * 6 ,
    'SAME_bias': [0.0378, 0.1418, -0.2974, 0.0039, 0.0555, -0.1933, -0.0114, 0.0639, -0.1211, 0.0215, 0.1563, -0.2637, -0.0098, 0.0296, -0.0607, 0.0086, 0.0632, -0.2299]
}

df = pd.DataFrame.from_dict(data)
df['model_config'] = df['model_name'] + " - " + df['embedding_config']
df.head()
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="model_config", y="SAME_bias", hue="gender", alpha=.6, height=6
)

g.set_axis_labels("Model Configuration", "SAME Bias score")
g.set(title="SAME Bias Score across HF Model & Embedding configuration")
g.fig.set_figwidth(14)
g.fig.set_figheight(8)
plt.show()
```
